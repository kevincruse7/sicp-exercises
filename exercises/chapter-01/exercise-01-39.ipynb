```racket
(define (tan-cf x k)
  (define (iter i acc)
    (if (= i 0)
        acc
        (iter (- i 1)
              (/ (* x x)
                 (- (- (* i 2) 1)
                    acc)))))
  (/ (iter k 0) x))
```

```racket
(tan-cf 5.0 10)
```
