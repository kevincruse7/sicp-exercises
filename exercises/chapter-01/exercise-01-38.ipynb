```racket
(define (cont-frac n d k)
  (define (helper i)
    (if (> i k)
        0
        (/ (n i)
           (+ (d i)
              (helper (+ i 1))))))
  (helper 1))
```

```racket
(+ (cont-frac (lambda (i) 1)
              (lambda (i)
                (if (= (remainder i 3) 2)
                    (* 2 (/ (+ i 1) 3))
                    1))
              10)
   2.0)
```
