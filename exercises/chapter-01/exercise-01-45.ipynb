From experimentation, it appears you need $\lfloor\log_{2}{n}\rfloor$ applications of `average-damp` to compute $n$-th
roots.

```racket
(define (compose f g)
  (lambda (x)
    (f (g x))))

(define (repeated f n)
  (if (= n 1)
      f
      (compose f (repeated f (dec n)))))

(define (nth-root n x)
  (fixed-point ((repeated average-damp (floor (log n 2)))
                (lambda (y) (/ x (expt y (dec n)))))
               1.0))
```

```racket
(nth-root 7 128)
```

```racket
(nth-root 8 256)
```

```racket
(nth-root 16 65536)
```
