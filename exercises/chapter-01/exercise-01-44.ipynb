```racket
(define (compose f g)
  (lambda (x)
    (f (g x))))

(define (repeated f n)
  (if (= n 1)
      f
      (compose f (repeated f (dec n)))))

(define (smooth f)
  (define dx 0.00001)
  (lambda (x)
    (/ (+ (f (- x dx))
          (f x)
          (f (+ x dx)))
       3)))
```

```racket
((smooth cube) 5)
```

```racket
(define (n-fold-smooth f n)
  ((repeated smooth n) f))
```

```racket
((n-fold-smooth cube 5) 5)
```
