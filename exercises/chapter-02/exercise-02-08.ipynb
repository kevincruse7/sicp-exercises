Given intervals $a$ and $b$, one can determine the lower bound of $a - b$ by subtracting the upper bound of $b$ from the lower bound of $a$. Likewise, the upper bound can be found by subtracting the lower bound of $b$ from the upper bound of $a$.

```racket
(define (sub-interval a b)
  (make-interval (- (lower-bound a) (upper-bound b))
                 (- (upper-bound a) (lower-bound b))))
```

```racket
(print-interval (sub-interval (make-interval 1 5)
                              (make-interval 2 6)))
```

```racket
(print-interval (sub-interval (make-interval 1 5)
                              (make-interval -6 -2)))
```
