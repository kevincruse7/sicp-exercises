```racket
(define (reverse sequence)
  (accumulate (lambda (x y) (append y (list x))) nil sequence))

(display (reverse (list 1 2 3 4 5)))
```

```racket
(define (reverse sequence)
  (fold-left (lambda (x y) (cons y x)) nil sequence))

(display (reverse (list 1 2 3 4 5)))
```
