```racket
(define (div-interval x y)
  (if (and (<= (lower-bound y) 0) (>= (upper-bound y) 0))
      (error "Cannot divide by interval spanning 0 -- DIV-INTERVAL")
      (mul-interval x
                    (make-interval (/ 1.0 (upper-bound y))
                                   (/ 1.0 (lower-bound y))))))
```

```racket
(div-interval (make-interval 1 5)
              (make-interval -1 2))
```
