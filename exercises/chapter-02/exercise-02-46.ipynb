```racket
(define (make-vect xcor ycor)
  (list xcor ycor))

(define (xcor-vect vect)
  (car vect))

(define (ycor-vect vect)
  (cadr vect))

(define vect1 (make-vect 1 3))
(display (xcor-vect vect1))
(newline)
(display (ycor-vect vect1))
```

```racket
(define (add-vect vect1 vect2)
  (make-vect (+ (xcor-vect vect1) (xcor-vect vect2))
             (+ (ycor-vect vect1) (ycor-vect vect2))))

(define vect2 (make-vect 6 5))
(display (add-vect vect1 vect2))
```

```racket
(define (sub-vect vect1 vect2)
  (make-vect (- (xcor-vect vect1) (xcor-vect vect2))
             (- (ycor-vect vect1) (ycor-vect vect2))))

(display (sub-vect vect1 vect2))
```

```racket
(define (scale-vect scale vect)
  (make-vect (* scale (xcor-vect vect))
             (* scale (ycor-vect vect))))

(display (scale-vect 2 vect1))
```
