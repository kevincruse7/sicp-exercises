```racket
(define (cc amount coin-values)
  (cond ((= amount 0) 1)
        ((or (< amount 0) (no-more? coin-values)) 0)
        (else
         (+ (cc amount
                (except-first-denomination coin-values))
            (cc (- amount
                   (first-denomination coin-values))
                coin-values)))))

(define (first-denomination coin-values)
  (car coin-values))

(define (except-first-denomination coin-values)
  (cdr coin-values))

(define (no-more? coin-values)
  (null? coin-values))

(define us-coins (list 50 25 10 5 1))
(define uk-coins (list 100 50 20 10 5 2 1 0.5))
```

```racket
(cc 100 us-coins)
```

```racket
(cc 100 uk-coins)
```

The order of the coins list does not matter, as regardless of the ordering, the tree recursive structure of `cc` breaks down into a sequence of equivalent leaf nodes that are added together. Since addition is commutative, the ordering of these additions does not change the result.
