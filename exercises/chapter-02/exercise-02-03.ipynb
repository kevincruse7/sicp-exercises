### Representation via opposing corners

```racket
(define (make-point x y)
  (cons x y))

(define (x-point point)
  (car point))

(define (y-point point)
  (cdr point))

(define (make-rectangle primary-corner opposing-corner)
  (cons primary-corner opposing-corner))

(define (primary-corner-rectangle rectangle)
  (car rectangle))

(define (opposing-corner-rectangle rectangle)
  (cdr rectangle))

(define (area-rectangle rectangle)
  (let ((primary-corner (primary-corner-rectangle rectangle))
        (opposing-corner (opposing-corner-rectangle rectangle)))
    (abs (* (- (x-point primary-corner)
               (x-point opposing-corner))
            (- (y-point primary-corner)
               (y-point opposing-corner))))))

(define (perimeter-rectangle rectangle)
  (let ((primary-corner (primary-corner-rectangle rectangle))
        (opposing-corner (opposing-corner-rectangle rectangle)))
    (double (+ (abs (- (x-point primary-corner)
                       (x-point opposing-corner)))
               (abs (- (y-point primary-corner)
                       (y-point opposing-corner)))))))

(define rectangle (make-rectangle (make-point 0 0)
                                  (make-point 10 10)))
```

```racket
(area-rectangle rectangle)
```

```racket
(perimeter-rectangle rectangle)
```

### Representation via a corner, a width, and a height

```racket
(define (make-rectangle primary-corner width height)
  (cons primary-corner (cons width height)))

(define (primary-corner-rectangle rectangle)
  (car rectangle))

(define (width-rectangle rectangle)
  (car (cdr rectangle)))

(define (height-rectangle rectangle)
  (cdr (cdr rectangle)))

(define (opposing-corner-rectangle rectangle)
  (let ((primary-corner (primary-corner-rectangle rectangle)))
    (make-point (+ (x-point primary-corner) (width-rectangle rectangle))
                (+ (y-point primary-corner) (height-rectangle rectangle)))))

(define rectangle (make-rectangle (make-point 0 0) 10 10))
```

```racket
(area-rectangle rectangle)
```

```racket
(perimeter-rectangle rectangle)
```
