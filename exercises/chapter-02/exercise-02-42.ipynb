```racket
(define empty-board nil)

(define (adjoin-position row col board)
  (if (null?
       (filter (lambda (position)
                 (and (= (car position) row)
                      (= (cadr position) col)))
               board))
      (cons (list row col) board)
      (error "Position already exists" row col)))

(define (safe? col board)
  (let ((candidates
         (filter (lambda (position)
                   (= (cadr position) col))
                 board)))
    (if (= (length candidates) 1)
        (let ((queen (car candidates)))
          (= (length
              (filter (lambda (position)
                        (or (= (car position) (car queen))
                            (= (cadr position) (cadr queen))
                            (let ((slope
                                   (/ (- (car position) (car queen))
                                      (- (cadr position) (cadr queen)))))
                              (or (= slope 1)
                                  (= slope -1)))))
                      board))
             1))
        (error "Zero or many queens in column" col))))

(define (queens board-size)
  (define (queen-cols k)
    (if (= k 0)
        (list empty-board)
        (filter
         (lambda (positions) (safe? k positions))
         (flatmap
          (lambda (rest-of-queens)
            (map (lambda (new-row)
                   (adjoin-position new-row k rest-of-queens))
                 (enumerate-interval 1 board-size)))
          (queen-cols (- k 1))))))
  (queen-cols board-size))

(define solutions (queens 8))
(display (length solutions))
```
