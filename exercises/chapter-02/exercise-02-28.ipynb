```racket
(define (fringe items)
  (if (null? items)
      nil
      (let ((item (car items)))
        (if (pair? item)
            (append (fringe item) (fringe (cdr items)))
            (cons item (fringe (cdr items)))))))

(define x (list (list 1 2) (list 3 4)))
```

```racket
(display (fringe x))
```

```racket
(display (fringe (list x x)))
```
