```racket
(define (make-segment start end)
  (list start end))

(define (start-segment segment)
  (car segment))

(define (end-segment segment)
  (cadr segment))

(define segment (make-segment (make-vect 1 1)
                              (make-vect 2 2)))

(display (start-segment segment))
(newline)
(display (end-segment segment))
```
