```racket
(define (subsets s)
  (if (null? s)
      (list nil)
      (let ((rest (subsets (cdr s))))
        (append rest
                (map (lambda (subset)
                       (cons (car s) subset))
                     rest)))))

(display (subsets (list 1 2 3)))
```

This procedure for generating super-sets works by generating the super-set for the given set $S$ without some element $x$, or $S - x$. Then, it copies the result and, to one of the copies, appends $x$ to each of the subsets. In doing so, we produce all possible subsets by appending subsets without $x$ to those with $x$.
