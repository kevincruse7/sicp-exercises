```racket
(define (make-rat n d)
  (let ((abs-n (abs n))
        (abs-d (abs d)))
    (let ((g (gcd abs-n abs-d)))
      (cons ((if (> (* n d) 0) + -) (/ abs-n g))
            (/ abs-d g)))))
```

```racket
(print-rat (make-rat 5 10))
```

```racket
(print-rat (make-rat -5 10))
```

```racket
(print-rat (make-rat 5 -10))
```

```racket
(print-rat (make-rat -5 -10))
```
