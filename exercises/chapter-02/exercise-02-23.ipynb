```racket
(define (for-each proc items)
  (define (helper)
    (proc (car items))
    (for-each proc (cdr items)))
  (if (null? items)
      true
      (helper)))
```

```racket
(for-each (lambda (x) (newline) (display x))
          (list 57 321 88))
```
