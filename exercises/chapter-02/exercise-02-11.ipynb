```racket
(define (make-interval a b)
  (cons a b))

(define (lower-bound interval)
  (car interval))

(define (upper-bound interval)
  (cdr interval))

(define (print-interval interval)
  (newline)
  (display "[")
  (display (lower-bound interval))
  (display ",")
  (display (upper-bound interval))
  (display "]"))

(define (mul-interval x y)
  (let ((x-low-neg? (< (lower-bound x) 0))
        (x-up-neg? (< (upper-bound x) 0))
        (y-low-neg? (< (lower-bound y) 0))
        (y-up-neg? (< (upper-bound y) 0)))
    (cond ((and x-low-neg? x-up-neg? y-low-neg? y-up-neg?)
           (make-interval (* (upper-bound x) (upper-bound y))
                          (* (lower-bound x) (lower-bound y))))
          ((and x-low-neg? x-up-neg? y-low-neg? (not y-up-neg?))
           (make-interval (* (lower-bound x) (upper-bound y))
                          (* (lower-bound x) (lower-bound y))))
          ((and x-low-neg? x-up-neg? (not y-low-neg?) (not y-up-neg?))
           (make-interval (* (lower-bound x) (upper-bound y))
                          (* (upper-bound x) (lower-bound y))))
          ((and x-low-neg? (not x-up-neg?) y-low-neg? y-up-neg?)
           (make-interval (* (upper-bound x) (lower-bound y))
                          (* (lower-bound x) (lower-bound y))))
          ((and x-low-neg? (not x-up-neg?) y-low-neg? (not y-up-neg?))
           (make-interval (min (* (lower-bound x) (upper-bound y))
                               (* (upper-bound x) (lower-bound y)))
                          (max (* (lower-bound x) (lower-bound y))
                               (* (upper-bound x) (upper-bound y)))))
          ((and x-low-neg? (not x-up-neg?) (not y-low-neg?) (not y-up-neg?))
           (make-interval (* (lower-bound x) (upper-bound y))
                          (* (upper-bound x) (upper-bound y))))
          ((and (not x-low-neg?) (not x-up-neg?) y-low-neg? y-up-neg?)
           (make-interval (* (upper-bound x) (lower-bound y))
                          (* (lower-bound x) (upper-bound y))))
          ((and (not x-low-neg?) (not x-up-neg?) y-low-neg? (not y-up-neg?))
           (make-interval (* (upper-bound x) (lower-bound y))
                          (* (upper-bound x) (upper-bound y))))
          (else
           (make-interval (* (lower-bound x) (lower-bound y))
                          (* (upper-bound x) (upper-bound y)))))))
```

```racket
(print-interval (mul-interval (make-interval -1 10)
                              (make-interval -3 -2)))
```
