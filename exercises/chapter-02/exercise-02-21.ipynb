```racket
(define (square-list items)
  (if (null? items)
      nil
      (cons (square (car items))
            (square-list (cdr items)))))
```

```racket
(display (square-list (list 1 2 3 4)))
```

```racket
(define (square-list items)
  (map square items))
```

```racket
(display (square-list (list 1 2 3 4)))
```
