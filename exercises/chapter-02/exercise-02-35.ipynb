```racket
(define (count-leaves t)
  (accumulate + 0 (map (lambda (node)
                         (if (pair? node)
                             (count-leaves node)
                             1))
                       t)))

(define x (cons (list 1 2) (list 3 4)))
```

```racket
(count-leaves x)
```

```racket
(count-leaves (list x x))
```
