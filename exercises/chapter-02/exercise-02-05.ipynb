Since 2 and 3 are prime numbers, they share no common factors. As such, given the product $2^{a} 3^{b}$, you can extract $a$ by repeatedly dividing the product by 2 until the remainder is not 0. The number of divisions it took is the value of $a$. Likewise, you can extract $b$ by repeatedly dividing the product by 3.

```racket
(define (cons a b)
  (* (expt 2 a)
     (expt 3 b)))

(define (car c)
  (define (iter c acc)
    (if (and (> c 0) (= (remainder c 2) 0))
        (iter (/ c 2) (+ acc 1))
        acc))
  (iter c 0))

(define (cdr c)
  (define (iter c acc)
    (if (and (> c 0) (= (remainder c 3) 0))
        (iter (/ c 3) (+ acc 1))
        acc))
  (iter c 0))
```

```racket
(car (cons 12 15))
```

```racket
(cdr (cons 12 15))
```
