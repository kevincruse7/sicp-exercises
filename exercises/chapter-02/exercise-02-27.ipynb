```racket
(define (deep-reverse items)
  (define (helper items acc)
    (if (null? items)
        acc
        (let ((item (car items)))
          (helper (cdr items)
                  (cons (if (pair? item)
                            (deep-reverse item)
                            item)
                        acc)))))
  (helper items nil))

(define x (list (list 1 2) (list 3 4)))
(display (deep-reverse x))
```
