```racket
(define (make-interval a b)
  (cons a b))

(define (lower-bound interval)
  (car interval))

(define (upper-bound interval)
  (cdr interval))
```

```racket
(define interval (make-interval 1 5))
(lower-bound interval)
```

```racket
(upper-bound interval)
```
